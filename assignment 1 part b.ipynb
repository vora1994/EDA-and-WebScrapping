Extract the content from the URL:

```python
response = requests.get(website_url)
page = str(BeautifulSoup(response.content))
```

Parsing Html:

```python
x=BeautifulSoup(page, "lxml")
```

Structure the content:

```python
print(x.prettify())
```

Find all the tags in the Html:

```python
x.find_all('a')
```

Filter out the external links from the list of all links scrapped

```python
lists=x.find_all('a', attrs={'href': re.compile("^https://")})
listsExtract URL from the list data:
```

Extract URL from the list data:

```python
links = []

for a in lists:
    links.append(a.get("href"))
```

Extract text(element) from the list data:

```python
text = []
for a in lists:
    text.append(a.string)
```

Funtion definition for getting URL response

```python
def statuscheck(url):
   #"to find status of website" 
    res = requests.head(url)
    if res.status_code == 200:
       return True 
    else:
       return False
```

Call to the function of status check. Update the last list to the URL

```python
status=[]
date=[]
for a in links:
    status.append(statuscheck(a))
    now = datetime.now()
    date.append(now.strftime("%Y-%m-%d %H:%M:%S"))
```
